```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Import and observe data

```python
ms = pd.read_csv('data/single_stock_analysis/microsoft.csv', index_col = 0)
display(ms.head())
display(ms.describe()) 
```

## Distribution of Log Returns

```python
# Calculate log-daily return
ms['logReturn'] = np.log(ms.Close).shift(-1) - np.log(ms.Close)
ms['Return'] = (ms.Close.shift(-1) - ms.Close) / ms.Close
ms.head() 
```

Calculate sample statistics: mean and variance approximation.

```python
mu = ms['logReturn'].mean()
sigma = ms['logReturn'].std(ddof=1) # Delta Degrees of Freedom. Divisor used for calculation is: N-ddof (as we use sample data)
```

Plot a histogram to show the distribution of log return of Microsoft's stock (We can expect it to be very close to normal distribution).

```python
from scipy.stats import norm

# Create an empty DataFrame: density
density = pd.DataFrame()
# Create a sample space (a set of possible values) for Log Returns: x
# np.arange(start, stop, step) - create an Array of values (possible LogReturn values given extreme ends and small step)
density['x'] = np.arange(ms['logReturn'].min()-0.01, ms['logReturn'].max()+0.01, 0.001) 
density['pdf'] = norm.pdf(density['x'], mu, sigma)
density['cdf'] = norm.cdf(density['x'], mu, sigma)

ms['logReturn'].hist(bins=50, figsize=(15, 8))
plt.plot(density['x'], density['pdf'], color='red')
plt.show()
```

Since Log Returns are very close to a normal distribution, we can estimate probability of losses using norm.cdf.

## Probability of a loss

Calculate the probability of the stock price will drop over a certain percentage in a day.

**Probability Density Function (PDF)** is used to specify the probability of the random variable falling within a particular range of values, as opposed to taking on any one value.

- This probability is given by the integral of this variable's PDF over that rangeâ€”that is, it is given by the area under the density function but above the horizontal axis and between the lowest and greatest values of the range. 

- The probability density function is nonnegative everywhere, and its integral over the entire space is equal to 1.

**Cumulative Distribution Function (CDF)** of a real-valued random variable X, or just distribution function of X, evaluated at x, is the probability that X will take a value less than or equal to x.

```python
plt.plot(density['x'], density['pdf'])
plt.title('Probability Density Function (PDF)')
plt.xlabel('Log returns (daily)')
plt.ylim(0)
plt.show()
```

```python
plt.plot(density['x'], density['cdf'])
plt.title('Cumulative Density Function (CDF)')
plt.xlabel('Log returns (daily)')
plt.ylim(0)
plt.show()
```

```python
# probability that the stock price of microsoft will drop over 0% in a day
prob_return1 = norm.cdf(0, mu, sigma)
print('The Probability of dropping over 0% in a day is ', prob_return1)
```

```python
# probability that the stock price of microsoft will drop over 5% in a day
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The Probability of dropping over 5% in a day is ', prob_return1)
```

---

Calculate the probability of the stock price will drop over a certain percentage **in a year**

We assume that daily returns are independent. This allow us to use a simple conversion from daily to yearly (or other periods).

```python
# drop over 40% in 220 days (A Year have a total of 220 trading days)
# Original mu and sigma were calculated using daily returns, thus we need to scale it:
mu220 = 220*mu
sigma220 = (220**0.5) * sigma
print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))
```

## Value at Risk (VaR)

```python
# Value at risk(VaR)
# ppf - percent point function (a.k.a quantile function)
# To calculate 95% VaR is the same as to find 5% quantile of a daily return.
VaR = norm.ppf(0.05, mu, sigma)
print('Single day VaR at 95% level is: ', VaR)
print('In other words, there is a 5% chance that daily return will be lower than: ', VaR)
```

```python
# Quatile 
# 5% quantile
print('5% quantile ', norm.ppf(0.05, mu, sigma))
# 95% quantile
print('95% quantile ', norm.ppf(0.95, mu, sigma))
```

## Confidence Interval

If a sample is a good representative of the population, the population mean should be close to sample mean.

**GOAL**: estimate population mean using confidence interval with lower and upper bounds. 

Suppose you have a normally distributed random variable (log Returns of a stock) and would like to calculate the probability of its value occurring in the interval of the [mean +/- .5 standard deviations]. How would you go about calculating the probability?

One method is to **standardize the random variable**. Standardization of a normally distributed random variable enables an analyst or researcher to determine with ease the probability associated with a range of values for that variable by using a standardized distribution table.

A **z-score** is a number that indicates how far above or below the mean a given score is in the distribution in the standard deviation units i.e. how many standard deviations a score is above or below the mean.

$Z = \frac{x-\mu}{\sigma}$

The **central limit theorem** essentially have the following characteristics:
- Mean of Sample is the same as the mean of the population.
- The standard deviation which is calculated is the same as the standard deviation of the population divided by the square root of the sample size.
- Standard deviation of the population can be unknown, but in practice we can replace it with sample standard deviation, given sample size is large enough.

Thus, when using a `sample mean`, `standard deviation` in the denominator is replaced by `standard error`.

$ Z = \frac{x - \mu}{\frac{s}{\sqrt{n}}} $

## Hypothesis testing

If we plot daily returns directly, we see up and down movements. It is not clear whether the average daily returns are positive, negative or equal to zero.

To answer this question we can use Hypothesis testing. In statistics, hypothesis testing can use sample information to test the validity of conjectures about these parameters.

```python
# Log return goes up and down during the period
ms['logReturn'].plot(figsize=(20, 8))
plt.axhline(0, color='red')
plt.show()
```

### Step 1: Set hypothesis (two tail test example)

Two tail test:

$H_0 : \mu = 0$

$H_1 : \mu \neq 0$

- H0 means the average stock return is 0
- H1 means the average stock return is not equal to 0

We begin our hypothesis testing by assuming the null is correct.

### Step 2: Calculate test statistic

In most practical research, the standard deviation for the population of interest is not known. In this case, the standard deviation  is replaced by the estimated standard deviation s, also known as the standard error. **Since the standard error is an estimate for the true value of the standard deviation, the distribution of the sample mean  is no longer normal.**

Instead, the sample mean follows the **t-distribution**. The t distribution is also described by its degrees of freedom. For a sample of size n, the t distribution will have n-1 degrees of freedom. The notation for a t distribution with k degrees of freedom is t(k). As the sample size n increases, the t distribution becomes closer to the normal distribution, since the standard error s approaches the true standard deviation $\sigma$ for large n.

$$ z = \frac{x - \mu}{\frac{\sigma}{\sqrt{n}}} $$

```python
sample_mean = ms['logReturn'].mean()
sample_std = ms['logReturn'].std(ddof=1)
n = ms['logReturn'].shape[0]

# if sample size n is large enough, we can use z-distribution, instead of t-distribtuion
# mu = 0 under the null hypothesis

# Standardization:
zhat = (sample_mean - 0)/(sample_std/n**0.5)
print(zhat)
```

If z-hat is different from 0 significantly, we can infer that this sample is not sampled for the population with mean equal to 0. Then, we can reject the null. How do we find the significance level? 

### Step 3: Set desicion criteria

If z-hat falls into a rejection region (into either of two extreme ends of distribution) - we reject the null.

Here, we should notice that z-hat is also possible to take values in rejection region even if the null is correct, and the mu equal to 0. This chance is equal to alpha 5%. In other words, we have a 5% chance to reject null wrongly. This is called a type 1 error, and the probability of a type 1 error is identical to the level of significance level.

```python
# confidence level 95%
alpha = 0.05

zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft  # z-distribution is symmetric 
print(zleft, zright)
```

### Step 4:  Make decision - shall we reject H0?

We can already see that a z-hat does not fall into extreme tails of a distribution

```python
print('At significant level of {}, shall we reject the null Hypothesis: {}'.format(alpha, zhat>zright or zhat<zleft))
```

### One tail test example
One tail test:

$H_0 : \mu \leq 0$

$H_a : \mu > 0$

```python
# step 2
sample_mean = ms['logReturn'].mean()
sample_std = ms['logReturn'].std(ddof=1)
n = ms['logReturn'].shape[0]

# if sample size n is large enough, we can use z-distribution, instead of t-distribtuion
# mu = 0 under the null hypothesis
zhat = (sample_mean - 0)/(sample_std/n**0.5)
print(zhat)
```

```python
# step 3
alpha = 0.05

zright = norm.ppf(1-alpha, 0, 1)
print(zright)
```

```python
# step 4
print('At significant level of {}, shall we reject null hypothesis: {}'.format(alpha, zhat>zright))
```

### An alternative method: p-value

p - probability of value taking a more extreme value than our observation in given sample?

If p is less than alpha which is a threshold, it means that the null is unlikely to be true.

```python
# step 3 (p-value)
p = 1 - norm.cdf(zhat, 0, 1)
print(p)
```

```python
# step 4
print('At significant level of {}, shall we reject the null hypothesis: {}'.format(alpha, p < alpha))
```
